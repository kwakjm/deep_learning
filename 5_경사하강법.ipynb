```python
#공부시간 x와 성적 y의 넘파이배열
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

#데이터의 분포를 그래프
plt.scatter(x,y)
plt.show()
```

```python
# 기울기와 절편 초기화

a=b=0
#학습률 초기화
lr = 0.03
#반복횟수 2000회
epochs = 2001

#1/n을 위한 x의 개수
n = len(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc

# 폰트 설정
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False


path = 'c:/Windows/Fonts/arialbd.ttf'
font_name = fm.FontProperties(fname= path).get_name()
rc('font',family =font_name)


for i in range(epochs):
    y_pred = a*x+b
    error = y-y_pred
    
    a_diff = (2/n)*sum(-x*(error))  #오차함수를 a로 편미분
    b_diff = (2/n)*sum(-(error))    #오차함수를 b로 편미분
    
    a = a - lr*a_diff   #학습률을 곱해 a를 업데이트
    b = b - lr*b_diff   #학습률을 곱해 b를 업데이트
    
    if i % 100 ==0:  #epochs가 100번 반복할때마다 출력
        print("epochs=%.f, 기울기=%.04f, 절편=%.04f" % (i,a,b))
        
        
# 앞서 구한 최종 a 값을 기울기, b값을 y 절편에 대입하여 그래프 그림
y_pred = a*x+b

#그래프 출력
plt.scatter(x,y)
plt.plot(x,y_pred, 'r')
plt.xlabel('Studying time')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

### 다중 선형 회귀

```python
# 다중 선형 회귀 독립변수가 여러개인경우

x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
x3 = np.array([82,93,91,97])

# 3D 그래프로 표현
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(x1,x2,y)
plt.show()
```

```python
# 다중 선형 회귀
a1,a2,b = 0,0,0
lr = 0.01
epochs = 2001
n=len(x1)


for i in range(epochs):
    y_pred = a1*x1+a2*x2+b
    error = y-y_pred
    
    a1_diff = (2/n)*sum(-x1*(error))  #오차함수를 a로 편미분
    a2_diff = (2/n)*sum(-x2*(error))  #오차함수를 a로 편미분
    b_diff = (2/n)*sum(-(error))    #오차함수를 b로 편미분
    
    a1 = a1 - lr*a1_diff   #학습률을 곱해 a를 업데이트
    a2 = a2 - lr*a2_diff   #학습률을 곱해 a를 업데이트
    b = b - lr*b_diff   #학습률을 곱해 b를 업데이트
    
    if i % 100 ==0:  #epochs가 100번 반복할때마다 출력
        print("epochs=%.f, 기울기1=%.04f, 기울기2=%0.04f, 절편=%.04f" % (i,a1,a2,b))
        
print("실제 점수:",y)
print("예측 점수:",y_pred)
```

```python
# 텐서플로우를 활용한 선형 회귀분석
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

model = Sequential() #모델생성 
model.add(Dense(1,input_dim =1, activation = 'linear')) #Dense layer 추가, input demension 1개 output demension 1개 
model.compile(optimizer= 'sgd', loss = 'mse') # 학습 컴파일 옵션

#모델 학습
model.fit(x,y, epochs = 2000)

plt.scatter(x,y)
plt.plot(x,model.predict(x), 'r')
plt.show()
```

```python
#임의의 시간을 집어넣어 점수를 예측하는 모델을 테스트
hour = 7
prediction = model.predict([hour])
print(hour, prediction)
```

```python
#다중선형 회귀분석 -> 텐서플로우

x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])

model = Sequential() #모델생성 
model.add(Dense(1,input_dim =2, activation = 'linear')) #Dense layer 추가, input demension 1개 output demension 1개 
model.compile(optimizer= 'sgd', loss = 'mse') # 학습 컴파일 옵션

#모델 학습
model.fit(x,y, epochs = 2000)
```

```python
#임의의 학습시간과 과외 시간을 집어넣어 점수 예측하는 모델 테스트
hour = 7
private_class = 4
prediction = model.predict([[hour,private_class]])
print(hour,private_class,prediction)
```
