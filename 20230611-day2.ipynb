```python
# 텐서플로우를 활용한 선형 회귀분석
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

model = Sequential()   # 모델 생성
model.add(Dense(1, input_dim=1, activation='linear'))  # layer 추가
model.compile(optimizer='sgd', loss='mse')  # 학습 컴파일 옵션

# 모델 학습
model.fit(x, y, epochs=2000)  # 반복은 2000회

# 그래프로 확인
plt.scatter(x,y)  # 원 데이터를 scatter 로 그림
plt.plot(x, model.predict(x), 'r')  # 예측값을 선으로 표기
plt.show()

```

```python
# 7시간을 공부했을때 성적을 예측
hour = [7, 5, 11]
model.predict(hour)
```

```python
# 다중 선형 회귀 분석 -> 텐서플로우로 
# x1 = np.array([2,4,6,8])
# x2 = np.array([0,4,2,3])
x = np.array([[2,0],[4,4],[6, 2],[8, 3]])
y = np.array([81,93,91,97])

model = Sequential()   # 모델 생성
model.add(Dense(10, input_dim=2, activation='relu'))  # layer 추가
model.add(Dense(1, activation='linear'))  # layer 추가
model.compile(optimizer='sgd', loss='mse')  # 학습 컴파일 옵션

# 모델 학습
model.fit(x, y, epochs=2000)  # 반복은 2000회
```

```python
# 7시간, 과외를 4회 했을때 성적을 예측
hour = 7
cnt = 4
model.predict([[hour, cnt]])
```

```python
# 로지스틱 회귀분석 
x = np.array([2,4,6,8,10,12,14])
y = np.array([0,0,0,1,1,1,1])

model = Sequential()
model.add(Dense(1, input_dim=1, activation='sigmoid'))  # 이진 분류이므로 시그모이드 함수 사용
model.compile(optimizer='sgd', loss='binary_crossentropy')
model.fit(x,y,epochs=5000)

# 그래프로 확인
plt.scatter(x,y)  # 원 데이터를 scatter 로 그림
plt.plot(x, model.predict(x), 'r')  # 예측값을 선으로 표기
plt.show()

# 공부시간이 7시간인 경우 합격 여부 확인
model.predict([7])
```

```python
# 다층 퍼셉트론으로 XOR 문제 해결
w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array([ 1, 1])
b1 = 3
b2 = -1
b3 = -1

# 퍼셉트론 함수 
def MLP(x, w, b):
    y = np.sum(x * w) + b
    if y <= 0:
        return 0
    else:
        return 1

# NAND 게이트 
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1)

# OR 게이트
def OR(x1, x2):
    return MLP(np.array([x1,x2]), w12, b2)

def AND(x1, x2):
    return MLP(np.array([x1,x2]), w2, b3)

def XOR(x1,x2):
    return AND(NAND(x1, x2), OR(x1,x2))
```

```python
# 폐암 수술 후 1년이내에 사망자와 생존자의 자료를 가지고 학습
# np.loadtxt("./data/ThoraricSurgery3.csv", delimiter='')
# 1. HIDDEN LAYER는 2개층 -> 처음 node수가 30개, 두번째 10개
# 2. 적절한 활성함수 사용
# 3. 손실함수와 옵티마이저를 sgd, adam을 사용해서 결과 비교 
 
data = np.loadtxt('./data/ThoraricSurgery3.csv', delimiter=',')
x = data[:, 0:-1]  # x 요소  , 마지막 자료는 target
y = data[:, -1]   # target 
```

```python
# 1. hidden layer 는 2개층 -> 처음 node수가 30개, 두번째 10개
# 2. 적절한 활성함수 사용'
def make_model(input_dim):
    model = Sequential()
    model.add(Dense(30, input_dim=input_dim, activation='relu'))  # 처음은 input_dim 필요
    # 두번째 layer부터는 input_dim 필요 없음, 이전 layer의 out node의 개수가 input_dim 수 
    model.add(Dense(10, activation='relu'))   
    model.add(Dense(1, activation='sigmoid'))
    
    return model

# 3. 손실함수와 옵티마이저를 sgd, adam을 사용해서 결과 비교 
model1 = make_model(x.shape[1])
model1.compile(loss='binary_crossentropy', optimizer='sgd',metrics=['accuracy'])

# 모델 학습
model1.fit(x,y,epochs=2000, verbose=False)
```

```python
print(model1.history.history['loss'][-1])  # 마지감 손실함수의 값 출력
# 데이터 예측
predict_result = model1.predict(x[-1].reshape(1,-1))
print(f"실제 값 :{y[-1]},  예측 값 : {predict_result }")
```

```python
model2 = make_model(x.shape[1])
model2.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

# 모델 학습
model2.fit(x,y,epochs=2000, verbose=False)
print(model2.history.history['loss'][-1])  # 마지감 손실함수의 값 출력
# 데이터 예측
predict_result = model2.predict(x[-1].reshape(1,-1))
print(f"실제 값 :{y[-1]},  예측 값 : {predict_result }")
```
