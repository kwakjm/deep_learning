```python
##폐암 수술 후 1년이내에 사망자와 생존자의 자료를 가지고 학습
## 1. Hidden layer는 2개층 -> 처음 node수가 30개, 두번째 10개
## 2. 적절한 활성함수 사용
## 3. 손실함수와 옵티마이저를 sgd, adam을 사용해서 결과 비교
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
#훈련 데이터 읽어옴
data_set = np.loadtxt('./data/ThoraricSurgery3.csv', delimiter=',')

X= data_set[:,0:16]
y = data_set[:,-1]
```

```python
# 모델 생성, 컴파일, 실행
model = Sequential()
model.add(Dense(30, input_dim=16, activation = 'relu')) # Layer 추가
model.add(Dense(10, activation = 'relu')) # Layer 추가
model.add(Dense(1,activation = 'sigmoid'))

# 모델 컴파일, 조건(손실함수, 최적화, 정확도, 손실율)
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

#모델 학습
history = model.fit(X,y, epochs=100 ,batch_size=16)
```

```python
print(history.history,'\n')
print(history.history['loss'],'\n')
print(history.history['accuracy'],'\n')
```
