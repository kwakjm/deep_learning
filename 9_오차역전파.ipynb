```python
##폐암 수술 후 1년이내에 사망자와 생존자의 자료를 가지고 학습
## 1. Hidden layer는 2개층 -> 처음 node수가 30개, 두번째 10개
## 2. 적절한 활성함수 사용
## 3. 손실함수와 옵티마이저를 sgd, adam을 사용해서 결과 비교
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
#훈련 데이터 읽어옴
data_set = np.loadtxt('./data/ThoraricSurgery3.csv', delimiter=',')

X= data_set[:,0:16]
y = data_set[:,-1]
```

```python
def make_model(input_dim):
    model =Sequential()
    model.add(Dense(30, input_dim=input_dim, activation = 'relu')) #처음은 input_dim 필요
    #두번째 layer부터는 iput_dim 필요없음 이전 layer의 out node의 개수가 input_dim 수
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense)
```

```python
# 모델 생성, 컴파일, 실행
model = Sequential()
model.add(Dense(30, input_dim=16, activation = 'relu')) # Layer 추가
model.add(Dense(10, activation = 'relu')) # Layer 추가
model.add(Dense(1,activation = 'sigmoid'))

# 모델 컴파일, 조건(손실함수, 최적화, 정확도, 손실율)
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

#모델 학습
history = model.fit(X,y, epochs=100 ,batch_size=16)
```

```python

print(history.history['accuracy'],'\n')
```

```python
# 모델을 사용하여 입력 데이터에 대한 예측 확률을 얻음
y_pred_prob = model.predict(X)

# 예측 확률을 클래스로 변환
y_pred = np.round(y_pred_prob)

# 예측 결과 확인
print(y_pred)
```
